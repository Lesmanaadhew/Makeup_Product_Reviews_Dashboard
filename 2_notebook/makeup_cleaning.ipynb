## **A. PRODUCT DATASET**

### 1. General Overview

```python
# product dataset
dataset_prod = pd.read_csv("../0_source_init/product_info.csv")
df_prod = dataset_prod.copy()
```

### 2. Filter `product_name` to include only makeup products

```python
print("Amount of products before filtering: ", len(df_prod['product_name'].unique()))
```

```python
filter_makeup_up = ['Foundation', 'Concealer', 'Powder', 'Blush', 'Bronzer', 'Highliter', 'Primer', 'Eyeshadow', 'Eyeliner', 'Mascara', 'Pencil', 'Eyebrow', 'Lip', 'Lipstick', 'Liner', 'Setting', 'Countour', 'BB', 'CC', 'Tinted', 'Makeup', 'Brush', 'Brow', 'Sponge', 'Pomade', 'Stain', 'Translucent', 'Blush', 'Compact', 'Powder', 'Plumper', 'Eyelashes', 'Finishing']
filter_makeup = filter_makeup_up + [x.lower() for x in filter_makeup_up]

df_prod = df_prod[df_prod['product_name'].str.contains('|'.join(filter_makeup))]
```

```python
print(f"Duplicated rows: {len(df_prod[df_prod.duplicated()])}")
```

```python
print("Amount of products after filtering: ", len(df_prod['product_name'].unique()))
```

### 3. Clean the boolean column

Columns with numeric data types and only 2 or 3 unique values indicate that they are boolean columns

```python
df_prod.info()
```

```python
df_prod.apply(lambda x: len(x.unique()))
```

```python
df_prod = df_prod.copy()

df_prod['limited_edition'] = df_prod['limited_edition'].astype(bool)
df_prod['new'] = df_prod['new'].astype(bool)
df_prod['online_only'] = df_prod['online_only'].astype(bool)
df_prod['out_of_stock'] = df_prod['out_of_stock'].astype(bool)
df_prod['sephora_exclusive'] = df_prod['sephora_exclusive'].astype(bool)
```

### 4. Clean the `ingredients` values

```python
df_prod['ingredients'].sample(10)
```

Replace the `\` values with `/`

```python
df_prod = df_prod.copy()

df_prod['ingredients_clean'] = df_prod['ingredients'].apply(lambda x: x.replace(r'\\', '/')
                                                      .replace('\\', '/')
                                                      .replace("'Cheek Pop:', ", '') if pd.notna(x) else x)
```

Delete all unnecessary values

```python
df_prod['ingredients_clean'] = df_prod['ingredients_clean'].astype(str).replace(r";", ",", regex=True).replace([r"[^\w.,\- ()]"], "", regex=True)
df_prod['ingredients_clean'] = df_prod['ingredients_clean'].replace(r", ", "', '", regex=True)
df_prod['ingredients_clean'] = "['" + df_prod['ingredients_clean'] + "']"

df_prod['ingredients_clean']
```

Replace all synonyms of `water` with `water`

```python
df_prod[df_prod['ingredients_clean'].str.contains('eau|Eau|aqua|Aqua|water|Water')]['ingredients_clean'].sample(10)
```

```python
list_water = ["Eau", "Aqua"]
list_water_2 = ["WaterWaterWater", "WaterWater"]

df_prod['ingredients_clean'] = df_prod['ingredients_clean'].replace(list_water, "Water", regex=True).replace(list_water_2, "Water", regex=True)

df_prod['ingredients_clean'] = df_prod['ingredients_clean'].replace(r"Water\s*\(Water\)?", "Water", regex=True).replace(r"Water',\s*'Water\)", "Water", regex=True)

df_prod[df_prod['ingredients_clean'].str.contains('Water')]['ingredients_clean']
```

```python
df_prod = df_prod.copy()

df_prod['ingredients_clean'] = df_prod['ingredients_clean'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### 5. Remove the non-makeup categories

```python
df_prod = df_prod[df_prod['primary_category'] != 'Fragrance']
df_prod = df_prod[df_prod['primary_category'] != 'Hair']
df_prod = df_prod[df_prod['primary_category'] != 'Bath & Body']
```

```python
filter_skincare = df_prod[(df_prod['primary_category'] == 'Skincare') & (df_prod['product_name'].str.contains('Mask|powder|Powder|treatment|Treatment|Sleeping|Sleeping|Overnight'))]['product_id'].tolist()

df_prod = df_prod[~df_prod['product_id'].str.contains('|'.join(filter_skincare))]
```
